The code below gives you some ideas on what instructive plots may be helpful

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

measurements = pd.read_csv("/content/data-subcortical_aseg_measure-mean_hemi-both.csv")
measurements["Age"] = [float(e) for e in measurements["Age"]]

plotcols = list(set(measurements.columns) - set(["Age", "Subject", "Sex"]))
```

Plot the volume measurements as a function of the age

```python
for col in plotcols:
    ax = sns.lmplot(x="Age", y=col, data=measurements, ci=95)
    ax.fig.subplots_adjust(top=.95)
    plt.title(col)
    plt.show()
```

Plot the same, but now consider different pattern for male and female subjects

```python
for col in plotcols:
    ax = sns.lmplot(x="Age", y=col, hue="Sex", data=measurements, ci=95)
    ax.fig.subplots_adjust(top=.95)
    plt.title(col)
    plt.show()
```
